write a program to count words 

````python
text = '''
hello my name is ahmed 
i like python 
python is cool 
i write python code eveyday 
'''
words  = text.split()
print(words)
print(dir(dict()))
````

````python
bag.update({'cup': 1})
bag
````

````python
dic  = {}
for word in words:
    if word in dic:
        dic[word] += 1 
    else:
        dic[word] = 1 
dic
````

````python
dic = {}
for word in words:
    dic[word] = dic.get(word, 0) + 1 
dic 
````

````python
dic = { word: words.count(word) for word in set(words)   }
````

````python
dic['i'] = 2 
dic
````
